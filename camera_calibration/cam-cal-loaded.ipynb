# OpenCV 활용 Camera Calibration [Load]

### import

```python
import matplotlib.pyplot as plt
import cv2, pickle
```

### test 이미지

```python
img = cv2.imread('test.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

### 미리 저장된 calibration 값을 load

```python
cam_cal = pickle.load( open( "cam-cal.pkl", "rb" ) )
mtx = cam_cal["mtx"]
dist = cam_cal["dist"]
img_size = cam_cal["imagesize"]
```

### 결과

```python
undist_img = cv2.undistort(img, mtx, dist, None, mtx)
plt.imshow(undist_img)
```
